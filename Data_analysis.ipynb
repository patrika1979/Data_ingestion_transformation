```
df_2017 = spark.read.format("com.crealytics.spark.excel").option("header","true").option("inferSchema", "true").load("/FileStore/tables/Base_2017.xlsx")
```

```
df_2017.printSchema()
```

```
df_2018 = spark.read.format("com.crealytics.spark.excel").option("header","true").option("inferSchema", "true").load("/FileStore/tables/Base_2018.xlsx")
```

```
df_2019 = spark.read.format("com.crealytics.spark.excel").option("header","true").option("inferSchema", "true").load("/FileStore/tables/Base_2019.xlsx")
```

```
df_2017.createOrReplaceTempView("vendas_2017")
```

```
df_2018.createOrReplaceTempView("vendas_2018")
```

```
df_2019.createOrReplaceTempView("vendas_2019")
```

```
%sql create or replace temp view vendas_consolidadas as select * from vendas_2017
UNION ALL
select * from vendas_2018
UNION ALL
select * from vendas_2019
```

```
df_2017.printSchema()
```

<h1> Vendas por ano e mÃªs

<h2> 2017

```
%sql select sum(qtd_venda) as valor,  year(to_date(data_venda, "m/d/yy")) as ano,  month(to_date(data_venda, "m/d/yy")) as mes from vendas_consolidadas where year(to_date(data_venda, "m/d/yy")) = 2017 group by ano, mes
```

<h2> 2019

```
%sql select sum(qtd_venda) as valor,  year(to_date(data_venda, "m/d/yy")) as ano,  month(to_date(data_venda, "m/d/yy")) as mes from vendas_consolidadas where year(to_date(data_venda, "m/d/yy")) = 2019 group by ano, mes
```

<h1> Vendas por marca e linha

```
%sql select sum(qtd_venda) as valor, marca,linha as mes from vendas_consolidadas where year(to_date(data_venda, "m/d/yy")) = 2017 group by marca,linha
```

```
%sql select sum(qtd_venda) as valor, marca,linha as mes from vendas_consolidadas where year(to_date(data_venda, "m/d/yy")) = 2019 group by marca,linha
```

<h1> Vendas por marca, ano e mÃªs

```
%sql select sum(qtd_venda) as valor, marca, year(to_date(data_venda, "m/d/yy")) as ano,  month(to_date(data_venda, "m/d/yy")) as mes from vendas_consolidadas where year(to_date(data_venda, "m/d/yy")) = 2017 group by marca,ano,mes
```

<h1> Vendas por linha, ano e mÃªs

```
%sql select sum(qtd_venda) as valor, linha, year(to_date(data_venda, "m/d/yy")) as ano,  month(to_date(data_venda, "m/d/yy")) as mes from vendas_consolidadas where year(to_date(data_venda, "m/d/yy")) = 2017  group by linha,ano,mes order by mes
```
